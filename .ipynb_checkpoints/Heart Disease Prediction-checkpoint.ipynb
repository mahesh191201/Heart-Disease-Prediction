## Train-Test Splitting

```python
from sklearn.model_selection import train_test_split

pred=df.drop("target", axis=1)
target=df["target"]

X_train, X_test, Y_train, Y_test=train_test_split(pred, target, test_size=0.20)
```

# Model Fitting

```python
from sklearn.metrics import accuracy_score
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()

model.fit(X_train, Y_train)

LR_pred=model.predict(X_test)
```

```python
LR_score=round(accuracy_score(LR_pred, Y_test)*100)

print("The score of Logistic Regression model is: "+str(LR_score)+"%")
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

model=GaussianNB()

model.fit(X_train, Y_train)

GNB_Pred=model.predict(X_test)

GNB_score=round(accuracy_score(GNB_Pred, Y_test)*100)

print("The score of Naive Bayes model is: "+str(GNB_score)+"%")
```

### K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=7)
model.fit(X_train, Y_train)
KNN_pred=model.predict(X_test)

KNN_score=round(accuracy_score(KNN_pred, Y_test)*100)
print("The accuracy of KNN is: "+str(KNN_score)+"%")
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

max_accuracy=0;

for x in range(2000):
    model=DecisionTreeClassifier(random_state=x)
    model.fit(X_train, Y_train)
    DT_pred=model.predict(X_test)
    curr_accuracy=round(accuracy_score(DT_pred, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
model=DecisionTreeClassifier(random_state=best_x)
model.fit(X_train, Y_train)
DT_pred=model.predict(X_test)

DT_score=round(accuracy_score(DT_pred, Y_test)*100)
print("The accuracy of Decision Tree is: "+str(DT_score)+"%")
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

max_accuracy=0;

for x in range(200):
    model=RandomForestClassifier(random_state=x)
    model.fit(X_train, Y_train)
    Y_pred_rf=model.predict(X_test)
    curr_accuracy=round(accuracy_score(Y_pred_rf, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
rf=DecisionTreeClassifier(random_state=best_x)
rf.fit(X_train, Y_train)
Y_pred_rf=rf.predict(X_test)

RF_score=round(accuracy_score(Y_pred_rf, Y_test)*100)
print("The accuracy of Random Forest is: "+str(RF_score)+"%")
```

### XG Boost

```python
import xgboost as xgb

xgb_model=xgb.XGBClassifier(objective="binary:logistic", random_state=42)
xgb_model.fit(X_train, Y_train)

Y_pred_xgb=xgb_model.predict(X_test)

XGB_score=round(accuracy_score(Y_pred_xgb, Y_test)*100)
print("The accuracy of XGBoost is: "+str(XGB_score)+"%")
```

### Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(11, activation='relu', input_dim=13))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=300)
```

```python
Y_pred_nn = model.predict(X_test)

rounded = [round(x[0]) for x in Y_pred_nn]

Y_pred_nn = rounded
NN_score = round(accuracy_score(Y_pred_nn,Y_test)*100)

print("The accuracy score achieved using Neural Network is: "+str(NN_score)+"%")
```

```python
scores = [LR_score,GNB_score,KNN_score,DT_score,RF_score,XGB_score,NN_score]
model = ["Logistic Regression","Naive Bayes","K-Nearest Neighbors","Decision Tree","Random Forest","XGBoost","Neural Network"]    

for i in range(len(model)):
    print("The score of "+ model[i]+" is: "+str(scores[i])+" %")
```
